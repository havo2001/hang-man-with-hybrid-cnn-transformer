```python
import numpy as np
from collections import defaultdict
from collections import Counter
import matplotlib.pyplot as plt
from create_training_examples import *
```

```python
# Load the dataset
data = np.load('dataset.npz', mmap_mode='r')
```

```python
# Distribution of the length of words in the dataset and the test set
toks = data['toks']
tokens_lengths = [int(np.where(tok == 28)[0][0]) for tok in toks] # Since the padding token is 28
tokens_lengths_counter = Counter(tokens_lengths)
print("Tokens lengths count:", tokens_lengths_counter)

# Distribution of length of the words in the test set
with open('final_test_words.txt', 'r') as f:
    test_words = f.read().splitlines()
test_words_lengths = [len(word) for word in test_words]
test_words_lengths_counter = Counter(test_words_lengths)
print("Test words lengths count:", test_words_lengths_counter)

 
# Plot the distributions of lengths of tokens and test words
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Plot the first histogram on the first axis (ax1)
ax1.hist(tokens_lengths, bins=range(1, 41), align='left', rwidth=0.8, alpha=0.7, color='blue')
ax1.set_xlabel('Length of tokens')
ax1.set_ylabel('Frequency')
ax1.set_title('Distribution of Length of Tokens')
ax1.set_xticks(range(1, 41))
ax1.tick_params(axis='x', rotation=90)
ax1.grid(axis='y')

# Plot the second histogram on the second axis (ax2)
ax2.hist(test_words_lengths, bins=range(1, 41), align='left', rwidth=0.8, alpha=0.7, color='orange')
ax2.set_xlabel('Length of Test Words')
ax2.set_ylabel('Frequency')
ax2.set_title('Distribution of Length of Test Words')
ax2.set_xticks(range(1, 41))
ax2.tick_params(axis='x', rotation=90)
ax2.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
# Distribution of labels in the dataset
# Load the dataset and count the occurrences of each label
data = np.load('dataset.npz', mmap_mode='r')
labels_count = Counter(data['labels'])
print("Labels count:", labels_count)
characters = [ID2CHAR[i] for i in labels_count.keys()]
plt.bar(characters, labels_count.values())
plt.xlabel('Labels')
plt.ylabel('Count')
plt.title('Distribution of Labels')
plt.xticks(rotation=90)
plt.show()
```
